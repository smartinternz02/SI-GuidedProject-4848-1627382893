```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("household_power_consumption.txt",sep = ';',header = 0)
dataset
```

```python
dataset.drop(dataset[dataset['Global_active_power'] == "?"].index, inplace = True)
dataset.drop(dataset[dataset['Global_reactive_power'] == "?"].index, inplace = True)
dataset.drop(dataset[dataset['Global_intensity'] == "?"].index, inplace = True)
dataset.drop(dataset[dataset['Sub_metering_1'] == "?"].index, inplace = True)
dataset.drop(dataset[dataset['Sub_metering_2'] == "?"].index, inplace = True)
dataset.drop(dataset[dataset['Sub_metering_3'] == "?"].index, inplace = True)
```

```python
Q1 = dataset.quantile(0.25)
Q1
```

```python
Q3 = dataset.quantile(0.75)
Q3
```

```python
IQR = Q3-Q1
IQR
```

```python
dataset_without_outlier = dataset[~((dataset<(Q1-1.5*IQR))| (dataset>(Q3+1.5*IQR))).any(axis = 1)]
```

```python
new_dataset = dataset_without_outlier
new_dataset.shape
```

```python
x=dataset.iloc[:,1:].values
y=dataset.iloc[:,0].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# Multilinear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
x_train
```

```python
y_train
```

```python
mr=LinearRegression()
```

```python
mr.fit(x_train,y_train)
```

```python
y_predict = mr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_predict,y_test)
```

```python
import pickle
filename = 'PCA_model.pkl'
pickle.dump(mr, open(filename, 'wb'))
```
